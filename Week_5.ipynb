```
#1
def day_of_the_week(number):
  if number==1:
    return "Wednesday"
  if number==2:
    return "Thursday"
  if number==3:
    return "Friday"
  if number==4:
    return "Saturday"
  if number==5:
    return "Sunday"
  if number==6:
    return "Monday"
  if number==7:
    return "Tuesday"
#2
def january_day(day_number):
  while day_number>7:
    day_number=day_number-7
  if day_number==1:
    return "Wednesday"
  if day_number==2:
    return "Thursday"
  if day_number==3:
    return "Friday"
  if day_number==4:
    return "Saturday"
  if day_number==5:
    return "Sunday"
  if day_number==6:
    return "Monday"
  if day_number==7:
    return "Tuesday"

print("The 19th day in January is", january_day(19))
#3
def number_to_day(num):
  while num>7:
    num=num-7
  if num==1:
    return "Wednesday"
  if num==2:
    return "Thursday"
  if num==3:
    return "Friday"
  if num==4:
    return "Saturday"
  if num==5:
    return "Sunday"
  if num==6:
    return "Monday"
  if num==7:
    return "Tuesday"

print(number_to_day(60))
#4


def month_day_to_number(month_num, day_num):
  if month_num==1:
    return day_num
  if month_num==2:
    return day_num+31
  if month_num==3:
    return day_num+60
  if month_num==4:
    return day_num+91
  if month_num==5:
    return day_num+121
  if month_num==6:
    return day_num+152
  if month_num==7:
    return day_num+182
  if month_num==8:
    return day_num+213
  if month_num==9:
    return day_num+244
  if month_num==10:
    return day_num+274
  if month_num==11:
    return day_num+305
  if month_num==12:
    return day_num+335
#5
def month_to_num(month):
  if month=="January":
    return 1
  if month=="February":
    return 2
  if month=="March":
    return 3
  if month=="April":
    return 4
  if month=="May":
    return 5
  if month=="June":
    return 6
  if month=="July":
    return 7
  if month=="August":
    return 8
  if month=="September":
    return 9
  if month=="October":
    return 10
  if month=="November":
    return 11
  if month=="December":
    return 12
#6
def what_day_of_the_week(month,day_num):
  month_num=month_to_num(month)
  num=month_day_to_number(month_num, day_num)
  return number_to_day(num)

print("November 26th, 2020 is a", what_day_of_the_week("November", 26))


  


    
```
